```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

np.random.seed(42)
df1 = pd.DataFrame({
    'Accept': np.random.randint(0, 100, size=100)  # Example random data
})
```

```python

plt.figure(figsize = (6,3))
plt.title("Acceptance Ratio")
plt.hist(df1["Accept"],bins = 10)
plt.show()
sns.histplot(df1['Accept'] , kde = True)
```

```python
#Create a pandas series of batsman1 scores
s1=[20,15,10,25,30,35,28,40,45,60]
scores1=pd.Series(s1)
scores1
```

```python
plt.figure(figsize=(6,2))
plt.title("Boxplot for batsman scores")
plt.xlabel("Scores")
plt.boxplot(scores1 ,vert=False)
```

```python
#Add extreme values  to scores and plot the boxplot
s2=[20,15,10,25,30,35,28,40,45,60,120,180]
scores2=pd.Series(s2)
print(scores2)
plt.figure(figsize=(6,2))
plt.title("Boxplot for batsman scores")
plt.xlabel("Scores")
plt.boxplot(scores2 ,vert=False)
```
