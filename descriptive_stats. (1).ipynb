```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

np.random.seed(42)
df1 = pd.DataFrame({
    'Accept': np.random.randint(0, 100, size=100)  # Example random data
})

```

```python

plt.figure(figsize = (6,3))
plt.title("Acceptance Ratio")
plt.hist(df1["Accept"],bins = 10)
plt.show()
sns.histplot(df1['Accept'] , kde = True)
```
